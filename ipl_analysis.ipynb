```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/matches.csv")
print(df.columns.tolist())
print(df.shape)
```

```
print(df.head())
print(df.isnull().sum())
```

```
wins=df['winner'].value_counts()
print(wins)
wins.plot(kind='bar',color='blue')
plt.title("Most match wins in IPL")
plt.xlabel("Team")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```
toss_match=df[df['toss_winner']==df['winner']]
print(f"Toss winner won match: {len(toss_match)} out of {len(df)}")
```

```
season_counts=df['season'].value_counts().sort_index()
season_counts.plot(kind='line',marker='o', color='orange')
plt.title("IPL matches per season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.tight_layout()
plt.show()
```
